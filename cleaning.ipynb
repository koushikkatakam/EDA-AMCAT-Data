### Numerical features after removing outliers

```python
# Assuming df is your DataFrame with the original data
# Apply outlier replacement method
numerical_cols = ['Salary', '10percentage', '12graduation', '12percentage', 
                  'collegeGPA', 'conscientiousness', 'agreeableness', 
                  'extraversion', 'nueroticism', 'openess_to_experience']

for col in numerical_cols:
    median_value = df[col].median()

    def replace_outliers_with_median(val):
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        if val < lower_bound or val > upper_bound:
            return median_value
        else:
            return val

    df[col] = df[col].apply(replace_outliers_with_median)

# Now plot box plots for the cleaned numerical columns using Seaborn
plt.figure(figsize=(18, 16))

# Iterate through each column and plot it
for i, col in enumerate(numerical_cols):
    # Create a subplot at position i+1 in a 2x5 grid
    plt.subplot(4, 3, i+1)
    
    # Plot a boxplot of the current column using Seaborn
    sns.boxplot(x=df[col])
    
    # Set the title for the subplot
    plt.title('Box plot for ' + col)
    
    # Remove x-axis ticks and labels
    plt.xticks([])
    
    # Remove x-axis grid lines
    plt.grid(False)

# Adjust layout and display the plot
plt.tight_layout()
plt.show()
```
