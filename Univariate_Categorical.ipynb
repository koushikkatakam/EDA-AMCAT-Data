### Employees with which Designation are more?

```python
df['Designation'] = df['Designation'].replace('systems engineer','system engineer')

categorical_columns['Designation'].value_counts().head(10)
```

```python
top_ten_designations = df['Designation'].value_counts().head(10)

# Create count plot using Seaborn
plt.figure(figsize=(20, 6))
ax = sns.countplot(x='Designation', data=df, order=top_ten_designations.index, palette='pastel')

# Set labels and title
ax.set_xlabel('Designation')
ax.set_ylabel('Count')
ax.set_title('Count Plot for Designation')

# Calculate percentages and add annotations
total = len(df)
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x() + p.get_width() / 2., height + 0.5, '{:.1f}%'.format((height / total) * 100), ha='center')

# Show the plot
# plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

- Most of the students are employed as software engineer with 13.5% of total students followed by system engineer and software developer so on
